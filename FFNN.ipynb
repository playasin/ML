```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
samples = 100000
up = [i for i in range(10)]
down = [9-i for i in range(10)]

data = [up if i%2 == 0 else down for i in range(samples)]
label = [[1] if i%2 == 0 else [0] for i in range(samples)]

#print(up)
#print(down)
#print(data[:10])
#for d, l in zip(data, label):
#    print(d, l)
```

```python
x = tf.placeholder(tf.float32, shape=[None, 10]) #x 데이터의 개수를 모를 때 None
y_ = tf.placeholder(tf.float32, shape=[None, 1]) #y 데이터의 개수를 모를 때 None
```

```python
layer1 = tf.layers.dense(x, units=10, name='11') #shitf+tab을 누르면 signature가 나옴
layer2 = tf.layers.dense(layer1, units=5, name='12')
layer3 = tf.layers.dense(layer2, units=7, name='13')
output = tf.layers.dense(layer3, units=1, name='14')

variables = tf.trainable_variables()
for v in variables:
    print(v)

saver = tf.train.Saver(variables)
```

```python
pred = tf.nn.sigmoid(output)
accuracy = tf.metrics.accuracy(y_, tf.round(pred))

loss = tf.losses.sigmoid_cross_entropy(y_, output)
train_op = tf.train.GradientDescentOptimizer(1e-2).minimize(loss)
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    sess.run(tf.local_variables_initializer()) #accuracy를 위한 변수 innitailize
    
    for i in range(1000):
        _, _loss = sess.run([train_op, loss], feed_dict={x: data, y_: label})
        
        if i%100 == 0:
            print(_loss)
        saver.save(sess, './log/ffnn_model')
        #test_data = [[i for i in range(10)], [9-i for i in range(10)]]
        
        _pred, _accuracy = sess.run([pred, accuracy], feed_dict={x: data, y_: label})
        print('loss: {}'.format(_loss))
        print("\nprediction")
        
        for d, p, a in zip(data[:10], _pred[:10], _accuracy[:10]):
            print('{}{}{}'.format(d, p, a))
        #print(_pred[:5])
    #plt.show()
```
