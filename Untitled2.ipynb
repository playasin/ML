```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tf.reset_default_graph()

samples = 1000
x = np.array([float(i)*0.01 for i in range(samples+1)], dtype=np.float32)
sin = np.sin(x)
plt.plot(sin)
plt.show()
```

```python
train_input = tf.reshape(sin[:-1], shape=[-1, 5, 1])
print(train_input)
```

```python
train_input = tf.unstack(train_input, axis=1) #axis=1로 unstack 한다는 것은 위에 shape(200, 5, 1)에서 5를 기준으로 array를 list로 만들어줌

for i in train_input:
    print(i)
```

```python
state_size = 15
#rnn_cell = tf.nn.rnn_cell.BasicRNNCell(num_units=state_size)
rnn_cell = tf.nn.rnn_cell.LSTMCell(num_units=state_size)
outputs, state = tf.nn.static_rnn(rnn_cell, train_input, dtype=tf.float32) #실제에서는 static_rnn을 사용하지 않음

outputs = tf.concat(outputs, axis=0) 

```

```python
#for o in outputs:
#    print('output shape {}'.format(o.shape))
print(outputs.shape)
```

```python
#Wo = tf.Variable(tf.truncated_normal(shape=[state_size, 1]))
#Bo = tf.Variable(tf.zeros(shape=[1.]))

#y = tf.matmul(outputs, Wo)

y = tf.layers.dense(outputs, 1) #Wo, Bo를 정의하고 matmul을 작성하나 dense layer나 같음
```

```python
print(y.shape)
```

```python
y = tf.reshape(y, shape=[5, -1])

label = tf.reshape(sin[1:], shape=[-1, 5])
label = tf.transpose(label)

loss = 0
for i in range(len(train_input)):
    loss += tf.losses.mean_squared_error(label[i], y[i])

train_op = tf.train.GradientDescentOptimizer(1e-5).minimize(loss)
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    _pred_before = sess.run(tf.reshape(tf.transpose(y), shape=[-1]))
    plt.plot(_pred_before)
    plt.show()
    
    for i in range(1000):
        _, _loss = sess.run([train_op, loss])
        if i%100 == 0:
            print('step: {}, loss: {}'.format(i, _loss))
            
    _pred_after = sess.run(tf.reshape(tf.transpose(y), shape=[-1]))
    plt.plot(_pred_after )
    plt.show()
        
    #    print(_loss)
```
